# Mã thay thế đơn ký tự (Monoalphabetic Substitution Cipher)

## Giới thiệu

Mã thay thế đơn ký tự (Monoalphabetic Substitution Cipher) là phương pháp mã hóa trong đó
mỗi ký tự trong bản rõ được thay thế bởi một ký tự khác trong bảng chữ cái.

Khác với các phương pháp như Additive hay Affine sử dụng công thức toán học,
phương pháp này sử dụng một bảng ánh xạ giữa các ký tự của plaintext và ciphertext.

Ví dụ:

Plaintext :  abcdefghijklmnopqrstuvwxyz  
Ciphertext:  noqatrbecfuhxdqgylkhvijmpzsw

## 1. Cơ sở lý thuyết

Trong Mã thay thế đơn ký tự (Monoalphabetic Substitution Cipher),
mỗi ký tự trong plaintext được ánh xạ tới một ký tự khác trong ciphertext.

Ví dụ bảng ánh xạ:

| Plaintext  | a | b | c | d | e | f | g | h | i | j | k | l | m | n | o | p | q | r | s | t | u | v | w | x | y | z |
|------------|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|---|
| Ciphertext | q | w | e | r | t | y | u | i | o | p | a | s | d | f | g | h | j | k | l | z | x | c | v | b | n | m |

#### Công thức mã hóa

C = f(P)

Trong đó:

- P là ký tự trong plaintext
- C là ký tự trong ciphertext
- f là một phép hoán vị của bảng chữ cái

#### Công thức giải mã

P = f⁻¹(C)

Trong đó:

- f⁻¹ là ánh xạ ngược từ ciphertext về plaintext
- mỗi ký tự ciphertext sẽ được thay bằng ký tự plaintext tương ứng trong bảng ánh xạ ngược

## 2. Ý tưởng thuật toán

#### Thuật toán mã hóa

1. Tạo bảng ánh xạ giữa plaintext và ciphertext.
2. Duyệt từng ký tự trong bản rõ.
3. Thay thế ký tự theo bảng ánh xạ.
4. Ghép các ký tự lại để tạo bản mã.

#### Thuật toán giải mã

1. Tạo bảng ánh xạ ngược từ ciphertext sang plaintext.
2. Duyệt từng ký tự trong bản mã.
3. Thay thế ký tự theo bảng ánh xạ ngược.
4. Ghép lại thành bản rõ.

## 3. Cài đặt thuật toán mã hóa

```python
def monoalphabetic_encrypt(text, key_map):
    result = ""

    for ch in text:
        if ch.isalpha():

            is_upper = ch.isupper()

            ch = ch.lower()
            enc = key_map[ch]

            if is_upper:
                enc = enc.upper()

            result += enc

        else:
            result += ch

    return result
```

## 4. Cài đặt thuật toán giải mã

```python
def monoalphabetic_decrypt(text, key_map):

    reverse_map = {v: k for k, v in key_map.items()}
    result = ""

    for ch in text:
        if ch.isalpha():

            is_upper = ch.isupper()

            ch = ch.lower()
            dec = reverse_map[ch]

            if is_upper:
                dec = dec.upper()

            result += dec

        else:
            result += ch

    return result
```

## 3. Giải thích thuật toán

Thuật toán sử dụng một bảng ánh xạ giữa các ký tự.

Ví dụ:

a → q  
b → w  
c → e  
d → r  
...

Khi mã hóa, mỗi ký tự trong plaintext sẽ được thay thế bằng
ký tự tương ứng trong bảng.

Giải mã chỉ cần thực hiện ánh xạ ngược lại.

## 6. Chạy thử chương trình

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
key = "qwertyuiopasdfghjklzxcvbnm"

key_map = dict(zip(alphabet, key))

text = input("Hay nhap dong text muon ma hoa:")

cipher = monoalphabetic_encrypt(text, key_map)
print("Dong text sau khi ma hoa:", cipher)

plain = monoalphabetic_decrypt(cipher, key_map)
print("Dong text sau khi giai ma:", plain)
```
