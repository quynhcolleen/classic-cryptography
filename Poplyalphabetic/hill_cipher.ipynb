# Mã Hill (Hill Cipher)

## 1. Giới thiệu

Hill Cipher là một hệ mật mã cổ điển dựa trên **đại số tuyến tính**.  
Thuật toán được đề xuất bởi **Lester S. Hill vào năm 1929**.

Khác với các phương pháp thay thế đơn giản (Caesar, Vigenère), Hill Cipher:

- Mã hóa **một khối nhiều ký tự cùng lúc**
- Sử dụng **ma trận khóa**
- Áp dụng **phép nhân ma trận trong modulo 26**

## 2. Cơ sở lý thuyết

Hill Cipher sử dụng **ma trận khóa n × n** để mã hóa văn bản.  
Mỗi chữ cái được ánh xạ thành số:

A = 0, B = 1, C = 2, ... , Z = 25

Plaintext được chia thành **các khối ký tự có độ dài n**, sau đó chuyển thành **vector số**.

## 3. Ý tưởng thuật toán

Để cài đặt Hill Cipher bằng Python, ta cần thực hiện một số thao tác toán học với ma trận.

Các bước chính của chương trình:

#### 1. Chuẩn hóa plaintext
   - Chuyển toàn bộ về chữ thường
   - Loại bỏ khoảng trắng

#### 2. Chia plaintext thành các block có độ dài **n**  
   - n là kích thước của ma trận khóa

#### 3. Chuyển mỗi ký tự thành số
   - a = 0, b = 1, ..., z = 25

#### 4. Thực hiện phép nhân ma trận
   - Nhân ma trận key với vector plaintext

#### 5. Lấy kết quả **mod 26**

#### 6. Chuyển kết quả trở lại thành ký tự để tạo ciphertext

Đối với giải mã:

1. Tính **ma trận nghịch đảo của key trong mod 26**
2. Nhân ciphertext với **ma trận nghịch đảo**
3. Lấy mod 26
4. Chuyển lại thành plaintext

Để thực hiện các bước này, ta cần xây dựng các hàm:

- Nhân ma trận
- Tính định thức
- Tính minor của ma trận
- Tìm nghịch đảo modulo
- Tính ma trận nghịch đảo
- Hàm mã hóa Hill
- Hàm giải mã Hill

## 4. Hàm nhân ma trận

```python
def multiply_matrix(A, B):
    n = len(A)
    p = len(A[0])
    m = len(B[0])

    result = [[0] * m for _ in range(n)]

    for i in range(n):
        for j in range(m):
            for k in range(p):
                result[i][j] += A[i][k] * B[k][j]

    return result
```

Hàm thực hiện phép **nhân hai ma trận A và B**.
Giả sử:
A có kích thước n × p  
B có kích thước p × m  
→ Khi đó kết quả sẽ là ma trận **n × m**.

## 5. Hàm lấy minor của ma trận

```python
def get_minor(matrix, i, j):
    return [row[:j] + row[j+1:] for row in (matrix[:i] + matrix[i+1:])]
```

Minor của phần tử (i, j) là **ma trận con** thu được khi:
- bỏ hàng i
- bỏ cột j

Hàm này được dùng khi:
- tính định thức
- tạo ma trận **cofactor** khi tính nghịch đảo ma trận

## 6. Hàm tính định thức

```python
def determinant(matrix):
    n = len(matrix)

    if n == 1:
        return matrix[0][0]

    if n == 2:
        return matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]

    det = 0
    for c in range(n):
        det += ((-1)**c) * matrix[0][c] * determinant(get_minor(matrix,0,c))

    return det
```

Hàm tính **định thức của ma trận vuông**.

**Trường hợp đặc biệt:**

- Ma trận 1x1 (là giá trị phần tử duy nhất)
- Ma trận 2x2 (nhân chéo rồi trừ)

Với ma trận lớn hơn, định thức được tính bằng **khai triển Laplace theo hàng đầu tiên**.

## 7. Hàm tìm nghịch đảo modulo

```python
def mod_inverse(a, m):
    a %= m

    for x in range(1, m):
        if (a * x) % m == 1:
            return x

    return None
```

Hàm tìm **nghịch đảo của a trong modulo m**.
Tức là tìm số x sao cho:
##### a × x mod m = 1

Trong Hill Cipher:
##### m = 26

Nếu không tìm được x thì **ma trận key không khả nghịch**.

## 8. Hàm tính nghịch đảo ma trận

```python
def matrix_inverse(matrix):
    n = len(matrix)

    det = determinant(matrix) % 26
    det_inv = mod_inverse(det, 26)

    if det_inv is None:
        print("Key khong kha nghich mod 26")
        exit()

    cofactors = []

    for r in range(n):
        row = []
        for c in range(n):
            minor = get_minor(matrix, r, c)
            row.append(((-1)**(r+c)) * determinant(minor))
        cofactors.append(row)

    adjugate = list(map(list, zip(*cofactors)))

    inverse = []

    for r in range(n):
        row = []
        for c in range(n):
            row.append((adjugate[r][c] * det_inv) % 26)
        inverse.append(row)

    return inverse
```

Để tìm **nghịch đảo ma trận trong mod 26**, ta thực hiện:

1. Tính định thức của ma trận
2. Tìm nghịch đảo của định thức mod 26
3. Tạo ma trận cofactor
4. Chuyển vị để tạo **adjugate matrix (ma trận phụ hợp)**
5. Nhân với nghịch đảo định thức
6. Lấy mod 26

## 9. Cài đặt hàm mã hóa

```python
def hill_encrypt(plaintext, key):
    n = len(key)

    plaintext = plaintext.lower().replace(" ", "")

    if len(plaintext) % n != 0:
        plaintext += "x" * (n - len(plaintext) % n)

    ciphertext = ""

    for i in range(0, len(plaintext), n):
        block = plaintext[i:i+n]

        vector = [[ord(c) - ord('a')] for c in block]

        result = multiply_matrix(key, vector)

        for r in result:
            ciphertext += chr((r[0] % 26) + ord('a'))

    return ciphertext
```

#### Các bước mã hóa:

1. Chuẩn hóa plaintext
2. Chia thành block độ dài n
3. Chuyển ký tự → số
4. Nhân với ma trận key
5. Lấy mod 26
6. Chuyển kết quả thành ciphertext

## 10. Hàm giải mã Hill Cipher

```python
def hill_decrypt(ciphertext, key):
    n = len(key)

    key_inv = matrix_inverse(key)

    plaintext = ""

    for i in range(0, len(ciphertext), n):
        block = ciphertext[i:i+n]

        vector = [[ord(c) - ord('a')] for c in block]

        result = multiply_matrix(key_inv, vector)

        for r in result:
            plaintext += chr((r[0] % 26) + ord('a'))

    return plaintext
```

#### Quy trình giải mã:

1. Tính ma trận nghịch đảo của key
2. Chia ciphertext thành block
3. Chuyển thành vector số
4. Nhân với ma trận nghịch đảo
5. Lấy mod 26
6. Chuyển lại thành plaintext

## 11. Chạy thử chương trình

```python
plaintext = input("Nhap plaintext: ")

n = int(input("Nhap kich thuoc key (n): "))

print("\nNhap ma tran key (theo hang):")

key = []

for i in range(n):
    row = list(map(int, input(f"Hang {i+1}: ").split()))

    if len(row) != n:
        print("Sai kich thuoc")
        exit()

    key.append(row)

print("\nKey:")
for r in key:
    print(r)

cipher = hill_encrypt(plaintext, key)

print("\nCiphertext:", cipher)

plain = hill_decrypt(cipher, key)

print("Plaintext (sau khi giai ma):", plain)
```
