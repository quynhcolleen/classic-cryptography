# Mã hóa cộng (Additive Cipher / Caesar Cipher)

## Giới thiệu

Additive Cipher (còn gọi là Caesar Cipher) là một trong những phương pháp mã hóa cổ điển đơn giản nhất trong mật mã học.

Ý tưởng chính của thuật toán là dịch chuyển mỗi ký tự trong bản rõ một số vị trí cố định trong bảng chữ cái.  
Số bước dịch chuyển này được gọi là **khóa (k)**.

Ví dụ nếu k = 3:

A → D  
B → E  
C → F  

Phương pháp này từng được sử dụng bởi Julius Caesar để gửi thông điệp bí mật trong quân đội La Mã.

## 1. Cơ sở lý thuyết

Trong Additive Cipher, mỗi ký tự trong bản rõ sẽ được chuyển thành một vị trí trong bảng chữ cái và dịch chuyển thêm k vị trí.

Công thức mã hóa:

E(x) = (x + k) mod 26

Trong đó:

- x là vị trí của ký tự trong bảng chữ cái
- k là khóa dịch chuyển
- mod 26 đảm bảo kết quả luôn nằm trong phạm vi bảng chữ cái

## 2. Ý tưởng thuật toán

Các bước thực hiện thuật toán:

1. Nhập chuỗi văn bản cần mã hóa.
2. Nhập giá trị khóa k.
3. Duyệt từng ký tự trong chuỗi.
4. Nếu ký tự là chữ thường (a–z):
   - Chuyển ký tự sang số bằng hàm ord()
   - Thực hiện phép dịch chuyển theo khóa k
   - Chuyển lại thành ký tự bằng hàm chr()
5. Nếu ký tự là chữ hoa (A–Z) thực hiện tương tự.
6. Nếu ký tự không phải chữ cái thì giữ nguyên.

## 3. Cài đặt thuật toán

```python
def additive_cipher(text, k):
    result = ""
    for ch in text:
        if ch.islower():
            result += chr((ord(ch) - ord("a")) + k % 26 + ord("a"))
        elif ch.isupper():
            result += chr((ord(ch) - ord("A")) + k % 26 + ord("A"))
        else:
            result += ch
    return result
```

## 4. Giải thích thuật toán

Thuật toán sử dụng hai hàm quan trọng trong Python:

- `ord()` : chuyển ký tự thành mã ASCII
- `chr()` : chuyển mã ASCII trở lại ký tự

Ví dụ:

ord('a') = 97  
ord('b') = 98  

Để đưa chữ cái về phạm vi 0–25, ta thực hiện:

ord(ch) - ord('a')

Sau đó thực hiện phép dịch chuyển theo khóa k:

(x + k) mod 26

Cuối cùng chuyển lại thành ký tự:

chr(value + ord('a'))

Điều này đảm bảo ký tự luôn nằm trong bảng chữ cái từ a đến z.

Đối với chữ hoa, thuật toán thực hiện tương tự nhưng sử dụng:

ord('A')

## 5. Chạy thử chương trình

```python
text = input("Hay nhap dong text muon ma hoa: ")
k = int(input("Hay nhap gia tri k: "))

cipher = additive_cipher(text, k)

print("Dong text sau khi ma hoa:", cipher)
```
