# Mã Vigenère (Vigenère Cipher)

## Giới thiệu

Mã Vigenere (Vigenere Cipher) là một phương pháp mã hóa thuộc nhóm
mã thay thế đa ký tự (polyalphabetic substitution cipher).

Khác với các hệ mã thay thế đơn ký tự như Caesar Cipher,
mã Vigenere sử dụng một **từ khóa (keyword)** để tạo ra
một chuỗi khóa lặp lại.

Nhờ đó mỗi ký tự trong plaintext có thể được mã hóa
bằng các giá trị dịch chuyển khác nhau, làm tăng
mức độ bảo mật so với các hệ mã cổ điển đơn giản.

## 1. Cơ sở lý thuyết

#### Giả sử

P = P₁ P₂ P₃ ... (plaintext)

C = C₁ C₂ C₃ ... (ciphertext)

K = (k₁, k₂, k₃, ..., kₘ, k₁, k₂, ...)

Trong đó:

- Pᵢ : giá trị của ký tự plaintext
- Cᵢ : giá trị của ký tự ciphertext
- kᵢ : giá trị của ký tự trong keyword

#### Công thức mã hóa

Cᵢ = (Pᵢ + kᵢ) mod 26

#### Công thức giải mã

Pᵢ = (Cᵢ − kᵢ) mod 26

## 2. Ý tưởng thuật toán

### Thuật toán mã hóa

1. Chuyển plaintext thành chữ thường.
2. Chuyển mỗi ký tự thành số từ 0–25.
3. Chuyển keyword thành chuỗi số tương ứng.
4. Lặp lại keyword cho đến khi có độ dài bằng plaintext.
5. Với mỗi ký tự:

   Cᵢ = (Pᵢ + kᵢ) mod 26

6. Chuyển kết quả thành ký tự.

### Thuật toán giải mã

1. Chuyển ciphertext thành số.
2. Lặp lại keyword giống khi mã hóa.
3. Với mỗi ký tự:

   Pᵢ = (Cᵢ − kᵢ) mod 26

4. Chuyển kết quả thành ký tự plaintext.

## 3. Cài đặt thuật toán mã hóa

```python
def vigenere_encrypt(text, key):
    result = ""
    key = key.lower()
    key_index = 0

    for ch in text:
        if ch.isalpha():
            p = ord(ch.lower()) - ord('a')
            k = ord(key[key_index % len(key)]) - ord('a')

            c = (p + k) % 26
            if ch.isupper():
                result += chr(c + ord('A'))
            elif ch.islower():
                result += chr(c + ord('a'))
                
            key_index += 1
        else:
            result += ch

    return result
```

## 4. Cài đặt thuật toán giải mã

```python
def vigenere_decrypt(cipher, key):
    result = ""
    key = key.lower()
    key_index = 0

    for ch in cipher:
        if ch.isalpha():
            c = ord(ch.lower()) - ord('a')
            k = ord(key[key_index % len(key)]) - ord('a')

            p = (c - k) % 26
            if ch.isupper():
                result += chr(p + ord('A'))
            elif ch.islower():
                result += chr(p + ord('a'))

            key_index += 1
        else:
            result += ch

    return result
```

## 5. Giải thích thuật toán

### Hàm mã hóa

Hàm `vigenere_encrypt()` thực hiện các bước:

1. Chuyển keyword thành chữ thường.
2. Với mỗi ký tự trong plaintext:
   - Chuyển ký tự thành số từ 0–25.
   - Lấy giá trị tương ứng trong keyword.
3. Áp dụng công thức:

   Cᵢ = (Pᵢ + kᵢ) mod 26

4. Chuyển kết quả trở lại ký tự.

### Hàm giải mã

Hàm `vigenere_decrypt()` thực hiện:

1. Lặp lại keyword giống quá trình mã hóa.
2. Với mỗi ký tự ciphertext:

   Pᵢ = (Cᵢ − kᵢ) mod 26

3. Chuyển kết quả về plaintext.

## 6. Chạy thử chương trình

```python
text = input("Hay nhap dong text muon ma hoa:")
key = input("Hay nhap keyword:")

key_index = 0
keystream = []
keystream_with_value = []

for t in text:
    if t.isalpha():
        keystream.append(key[key_index % len(key)])
        keystream_with_value.append(ord(key[key_index % len(key)].lower()) - ord ('a'))
        key_index += 1
        
print("")
print("Keystream (text):", keystream)
print("Keystream (gia tri):", keystream_with_value)
print("")

cipher = vigenere_encrypt(text, key)

print("Ciphertext:", cipher)

plain = vigenere_decrypt(cipher, key)

print("Giai ma:", plain)
```
