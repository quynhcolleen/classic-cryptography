# Mã hóa Affine (Affine Cipher)

## Giới thiệu

Mã hóa Affine (Affine Cipher) là một phương pháp mã hóa cổ điển trong mật mã học. Thuật toán này kết hợp hai phép biến đổi tuyến tính:

- phép nhân
- phép cộng

Affine Cipher có thể xem là sự kết hợp của:

Multiplicative Cipher + Additive Cipher

Nhờ đó mức độ bảo mật cao hơn so với từng phương pháp riêng lẻ.

## 1. Cơ sở lý thuyết

Trong Affine Cipher, mỗi ký tự được chuyển thành vị trí trong bảng chữ cái.

#### Công thức mã hóa:

E(x) = (a × x + b) mod 26

#### Công thức giải mã:

D(y) = a⁻¹ × (y − b) mod 26

#### Trong đó:

- x là vị trí ký tự trong bản rõ
- y là vị trí ký tự trong bản mã
- a là khóa nhân
- b là khóa cộng
- a⁻¹ là nghịch đảo của a theo modulo 26

#### Điều kiện của khóa:

gcd(a, 26) = 1

## 2. Ý tưởng thuật toán

#### Thuật toán mã hóa

1. Nhập chuỗi văn bản cần mã hóa.
2. Nhập hai khóa a và b.
3. Duyệt từng ký tự trong chuỗi.
4. Chuyển ký tự sang vị trí trong bảng chữ cái.
5. Tính (a × x + b) mod 26.
6. Chuyển kết quả trở lại ký tự.

#### Thuật toán giải mã

1. Tìm nghịch đảo của a theo modulo 26.
2. Duyệt từng ký tự trong bản mã.
3. Chuyển ký tự sang vị trí trong bảng chữ cái.
4. Tính a⁻¹ × (y − b) mod 26.
5. Chuyển kết quả trở lại ký tự ban đầu.

## 3. Cài đặt thuật toán mã hóa

```python
def affine_encrypt(text, a, b):
    result = ""
    for ch in text:
        if ch.islower():
            result += chr((a * (ord(ch) - ord('a')) + b) % 26 + ord('a'))
        elif ch.isupper():
            result += chr((a * (ord(ch) - ord('A')) + b) % 26 + ord('A'))
        else:
            result += ch
    return result
```

## 4. Cài đặt thuật toán giải mã

```python
def mod_inverse(a, m):
    for i in range(1, m):
        if (a * i) % m == 1:
            return i
    return None


def affine_decrypt(text, a, b):
    result = ""
    a_inv = mod_inverse(a, 26)

    for ch in text:
        if ch.islower():
            result += chr((a_inv * ((ord(ch) - ord('a')) - b)) % 26 + ord('a'))
        elif ch.isupper():
            result += chr((a_inv * ((ord(ch) - ord('A')) - b)) % 26 + ord('A'))
        else:
            result += ch

    return result
```

## 5. Giải thích thuật toán

#### Affine Cipher thực hiện hai bước biến đổi:

1. Nhân vị trí ký tự với a
2. Cộng thêm b

Sau đó lấy mod 26 để đảm bảo kết quả nằm trong bảng chữ cái.

#### Quá trình giải mã thực hiện ngược lại:

1. Trừ b
2. Nhân với nghịch đảo của a
3. Lấy mod 26

## 6. Chạy thử chương trình

```python
text = input("Hay nhap dong text muon ma hoa: ")

a = int(input("Nhap gia tri a: "))
b = int(input("Nhap gia tri b: "))

cipher = affine_encrypt(text, a, b)
print("Dong text sau khi ma hoa:", cipher)

plain = affine_decrypt(cipher, a, b)
print("Dong text sau khi giai ma:", plain)
```
