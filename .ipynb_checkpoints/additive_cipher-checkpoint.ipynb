# Additive Cipher (Mã cộng)

Additive cipher là một phương pháp mã hóa cổ điển trong mật mã học. Ý tưởng của nó là **dịch mỗi ký tự trong bản rõ đi một số vị trí cố định trong bảng chữ cái**.

Công thức toán học:

C = (P + k) mod 26

Trong đó:

- P: vị trí của ký tự trong bảng chữ cái
- k: khóa mã hóa
- C: ký tự sau khi mã hóa
- 26: số chữ cái trong bảng alphabet

Ví dụ:

Plaintext: HELLO  
k = 3

Sau khi mã hóa:

H → K  
E → H  
L → O  
L → O  
O → R

Kết quả: **KHOOR**

Nếu vượt quá 'z' hoặc 'Z' thì ký tự sẽ quay vòng về đầu bảng chữ cái.

## Công thức toán học

Thuật toán Additive Cipher được biểu diễn bằng công thức:

C = (P + k) mod 26

Trong đó:

- P: vị trí của ký tự trong bảng chữ cái
- k: khóa mã hóa
- C: ký tự sau khi mã hóa
- mod 26: phép toán chia dư cho 26

Bảng chữ cái tiếng Anh có 26 ký tự nên phép toán mod 26 giúp ký tự quay vòng khi vượt quá 'z' (hoặc 'Z').

## Nguyên lý hoạt động

Quy trình mã hóa gồm các bước:

1. Nhập chuỗi văn bản cần mã hóa (plaintext)
2. Nhập giá trị khóa k
3. Với mỗi ký tự trong chuỗi:
   - xác định vị trí của ký tự trong bảng chữ cái
   - cộng thêm giá trị k
   - nếu vượt quá bảng chữ cái thì quay vòng lại từ đầu
4. Ghép các ký tự đã mã hóa lại để tạo ciphertext

```python
def additive_cipher(text, k):
    result = ""
    k = k % 26

    for ch in text:
        if ch.islower():
            result += chr((ord(ch) - ord('a') + k) % 26 + ord('a'))
        elif ch.isupper():
            result += chr((ord(ch) - ord('A') + k) % 26 + ord('A'))
        else:
            result += ch

    return result
```

## Giải thích chương trình

Hàm `additive_cipher()` thực hiện việc mã hóa chuỗi theo thuật toán Additive Cipher.

Các bước xử lý:

1. `k = k % 26`

Đảm bảo giá trị dịch không vượt quá số ký tự trong bảng chữ cái.

Ví dụ:

k = 29

29 mod 26 = 3

Nên dịch 3 ký tự là đủ.

---

2. Kiểm tra ký tự là chữ hoa hay chữ thường

- `islower()` kiểm tra chữ thường
- `isupper()` kiểm tra chữ hoa

Điều này giúp giữ nguyên kiểu chữ ban đầu.

---

3. Chuyển ký tự thành mã ASCII

Hàm `ord()` chuyển ký tự sang số.

Ví dụ:

ord('a') = 97

---

4. Tính toán vị trí ký tự mới

Công thức:

(ord(ch) - ord('a') + k) % 26

Sau đó cộng lại `ord('a')` để quay về ASCII.

---

5. Chuyển ASCII thành ký tự

Hàm `chr()` chuyển số ASCII thành ký tự.

```python
text = input("Hay nhap dong text muon ma hoa: ")
k = int(input("Hay nhap gia tri k: "))

cipher = additive_cipher(text, k)

print("Dong text sau khi ma hoa:", cipher)
```

## Ưu điểm và nhược điểm

### Ưu điểm

- Thuật toán đơn giản
- Dễ hiểu và dễ cài đặt
- Phù hợp cho mục đích học tập

### Nhược điểm

- Rất dễ bị phá mã
- Chỉ có 26 khóa có thể thử
- Không đủ an toàn để sử dụng trong hệ thống bảo mật hiện đại
